```python
#Cargo el fichero 
df_climate1 =pd.read_csv("./GlobalLandTemperatures_GlobalLandTemperaturesByMajorCity_jgx.csv")
```

```python
# Averiguo el tamaño
df_climate1.shape
```

```python
#Visualizo el nombre de la columnas
print(df_climate1.columns)
```

```python
#Muestro las 6 primeras muestras
df_climate1.head(6)
```

```python
#Elimino la primera columna

del df_climate1['Unnamed: 0']
```

```python
#Elimino las muestras que no están completas
df_climate1.dropna(inplace=True) # Inplace, modifica el mismo dataframe al hacer dropna.
```

```python
#Averiguamos las últimas muestras del dataframe
df_climate1.tail(6)
```

```python
#La variable  df_climate3 apunta a un objeto de memoria diferente
df_climate3=df_climate1.copy() #Crea una copia de climate1_country
```

```python
#Filtramos las muesttras del dataframe de la ciudad de xian
df_climate3_country=df_climate3[(df_climate3['City']=='Xian')]
```

```python
#Convertimos la variable dt que era un string a un tipo datetime
df_climate3['dt'] = pd.to_datetime(df_climate3['dt'],format="%Y-%m-%d")
```

```python
#El método 'describe' permite obtener los estadísticos  que nos piden de las variables continuas
df_climate3.describe() #sólo incluye las variables numéricas

#df_climate3.describe(include='all') #Descripción de todas las varibales
```

```python
#Creamos df_climate4 y hacemos que el índice sea la variable dt.
df_climate4=df_climate3.copy()
df_climate4.set_index('dt',inplace=True) # Reliza la acción modificando el propio dataframe
```

```python
#Filtramos las muestras posteriores al 2 de enero de 1900.

df_climate4=df_climate4['1900-01-02':]
```
